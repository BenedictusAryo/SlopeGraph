# Slope Graph 


```python
import pandas as pd
import numpy as np
from plotSlope import slope
```

```python
df = pd.read_csv('data/EU_GDP_2007_2013.csv',index_col=0,na_values='-')
df.head()
```

```python
f = slope(df/1000, kind='interval',font_family='GillSans',
          font_size=30, height=40, width=50, dpi=200,
          title=u'European GPD until 2010 and forecasts at market prices (billions of Euro) source : EUROSTAT')
```
